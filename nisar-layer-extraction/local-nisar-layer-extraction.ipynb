````python
from pathlib import Path
from nisar.products.readers import open_product
from osgeo import gdal, osr
````

````python
# Change this to the path of a GCOV product on your system
granule_name = 'NISAR_L2_PR_GCOV_005_135_D_078_2005_QPDH_A_20251119T135024_20251119T135041_X05009_N_P_J_001.h5'
gcov_product = Path.home() / 'Data' / 'nisar' / granule_name
gcov_product
````

````python
gcov = open_product(gcov_product)
````

````python
print(gcov.frequencies)
print(gcov.polarizations)
````

````python
hhhh_dataset = gcov.getImageDataset(frequency='B', polarization='HHHH')
````
