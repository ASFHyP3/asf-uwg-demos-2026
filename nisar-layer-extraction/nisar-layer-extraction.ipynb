## Nisar GCOV Layer Extraction

`harmony-py` is the programatic interface for submitting harmony jobs
- Github: https://github.com/nasa/harmony-py
- Docs: https://harmony-py.readthedocs.io/en/latest

````python
%pip install harmony-py
````

````python
import harmony
````

````python
harmony_client = harmony.Client(env=harmony.Environment.PROD)
````

To submit a harmony job, you first need to setup a `.netrc` file with your Earthdata Login Credentials.

```
machine urs.earthdata.nasa.gov
    login <EDL Username>
    password <EDL Password>
```

Create a request to submit a harmony job. We specify the granule name, the layer we want (frequency B and HVHVH), and the output format (which will be a geotiff).

````python

request = harmony.Request(
    collection=harmony.Collection(id="NISAR_L2_GCOV_BETA_V1"),
    granule_name="NISAR_L2_PR_GCOV_010_164_D_076_2005_QPDH_A_20260120T140558_20260120T140633_X05010_N_F_J_001",
    variables=[
        "/science/LSAR/GCOV/grids/frequencyB/HVHV",
    ],
    format="image/tiff"
)


job_id = harmony_client.submit(request)
````

Wait for the harmony job to finish processing

````python
harmony_client.wait_for_processing(job_id)
````

````python
data = harmony_client.result_json(job_id)
data
````

````python
urls = list(harmony_client.result_urls(job_id))
urls
````

Download the resuts to our current directory

````python
results = harmony_client.download_all(job_id, directory='.', overwrite=True)
file_names = [f.result() for f in results]
file_names
````
