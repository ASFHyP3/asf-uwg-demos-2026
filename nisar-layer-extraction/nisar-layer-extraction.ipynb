## Nisar GCOV Layer Extraction

`harmony-py` is the programatic interface for submitting harmony jobs
- Github: https://github.com/nasa/harmony-py
- Docs: https://harmony-py.readthedocs.io/en/latest

````python
import harmony
````

````python
harmony_client = harmony.Client(env=harmony.Environment.PROD)
````

To submit a harmony job, you first need to setup a `.netrc` file with your Earthdata Login Credentials.

```
machine urs.earthdata.nasa.gov
    login <EDL Username>
    password <EDL Password>
```

````python

request = harmony.Request(
    collection=harmony.Collection(id="NISAR_L2_GCOV_BETA_V1"),
    granule_name="NISAR_L2_PR_GCOV_008_145_D_054_2005_QPDH_A_20251226T061551_20251226T061609_X05010_N_P_J_001",
    variables=[
        "/science/LSAR/GCOV/grids/frequencyB/VVVV",
    ],
)

job_id = harmony_client.submit(request)
````
